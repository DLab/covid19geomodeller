# Agregar Datos

## Agregar Filas
Agregar elementos (filas) al final del dataframe en el dataframe,conservando las mismas columnas existentes, se puede hacer de varias formas.
La forma más sencilla es la siguiente:  
```  dataframe.loc[i] = [data] ```   
Donde:
* **dataframe**: Es tu dataframe =)
* **i** es el número de fila que quieres modificar o es una más abajo de la última para agregar una nueva
* **data** es la data que quieres agregar, la cual tiene que tener la misma estructura del dataframe (la misma cantidad de columnas)

## Agregar Columnas
Para agregar una nueva columna existe el comando insert que tiene la siguiente estructura:
```datafframe.insert(loc,nombrecolumna,valores)```
Donde: 
* **dataframe**: Es tu dataframe =)
* **loc** es la posición donde se agregará la nueva columna
* **nombrecolumna** es el nombre de la columna
* **valores** es la data que quieres agregar, la cual tiene que tener la misma estructura del dataframe (la misma cantidad de filas)
