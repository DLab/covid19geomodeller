```python
import matplotlib.pyplot as plt
plt.plot(model.S[False], label='S')
plt.plot(model.S[True], label='Sv')
plt.plot(model.E[False], label='E')
plt.plot(model.E[True], label='Ev')
plt.plot(model.Im[False], label='Im')
plt.plot(model.Im[True], label='Imv')
plt.plot(model.Icr[False], label= 'Icr')
plt.plot(model.Icr[True], label = 'Icrv')
plt.plot(model.R[True] + model.R[False], label='R')
plt.plot(model.H[True] + model.H[False], label='H')
plt.plot(model.D[True] + model.D[False], label='D')
plt.legend(loc=0)
plt.show()
```
